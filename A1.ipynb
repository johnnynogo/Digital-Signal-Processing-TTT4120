```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd

Fs = 8000
F1_l = [1000, 3000, 6000]
T = 2
A = 1

for F1 in F1_l:
    N = Fs * T
    f1 = F1 / Fs
    n = np.arange(N)
    x = A * np.cos(2 * np.pi * f1 * n)

    sd.play(x, Fs)
    sd.wait()
```
